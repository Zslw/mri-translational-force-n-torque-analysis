```python
import pandas as pd
import os
import re

# List of July 4th files
july4_files = [
    'dataTorque/Data_Rod2_endN1-070425.csv',
    'dataTorque/Data_Rod2_endN2-070425.csv',
    'dataTorque/Data_Rod1_endN1-070425.csv',
    'dataTorque/Data_Rod3_endN2-070425.csv',
    'dataTorque/Data_Rod7_endN2-070425.csv',
    'dataTorque/Data_Rod7_endN1-070425.csv',
    'dataTorque/Data_Rod3_endN1-070425.csv',
    'dataTorque/Data_Rod1_endN2-070425.csv',
    'dataTorque/Data_Rod2_endN1-070425.csv',
    'dataTorque/Data_Rod5_endN2-070425.csv',
    'dataTorque/Data_Rod5_endN1-070425.csv'

]

# List of July 11th files

july11_files = [
    #'New_dataTorque/Data_Rod3_endN1-071125-upto14points.csv',
    #'New_dataTorque/Data_Rod3_endN2-071125.csv',
    'New_dataTorque/Data_Rod3_endN3-071125.csv',
    #'New_dataTorque/Data_Rod3_endN4-071125.csv',
    'New_dataTorque/Data_Rod3_90_endN1-071125.csv',
    'New_dataTorque/Data_Rod3_180_endN1-071125.csv',
    #'New_dataTorque/Data_Rod3_180_endN2-071125.csv',
    #'New_dataTorque/Data_Rod2_endN1-071125_discard.csv',
    'New_dataTorque/Data_Rod2_endN2-071125.csv',
    'New_dataTorque/Data_Rod2_90_endN1-071125.csv',
    'New_dataTorque/Data_Rod2_180_endN1-071125.csv',
    #'New_dataTorque/Data_Rod2_180_endN2-071125.csv'
]

compiled_data = []

# Parse metadata and extract values
for file_path in july11_files:
    fname = os.path.basename(file_path)
    rod_match = re.search(r'Rod(\d+)', fname)
    rod_id = f"Rod{rod_match.group(1)}" if rod_match else "Unknown"
    position_match = re.search(r'endN\d+', fname)
    position = position_match.group(0) if position_match else "Unknown"
    
    angle_match = re.search(r'_(\d{2,3})_', fname)
    if not angle_match:
        angle_match = re.search(r'_(\d{2,3})-', fname)
    angle = int(angle_match.group(1)) if angle_match else 0



    
    df = pd.read_csv(file_path, header=None)
    
    for i, row in df.iterrows():
        for idx, val in enumerate(row):
            compiled_data.append({
                "rod_id": rod_id,
                "position": position,
                "angle": angle,
                "measurement_index": i,
                "point_index": idx,
                "value": val
            })

# Convert to DataFrame and save
compiled_df = pd.DataFrame(compiled_data)
compiled_path = "compiled_torque_july11.dat"
compiled_df.to_csv(compiled_path, sep=",", index=False)
compiled_path
```

```python
import pandas as pd
import os
import re

# List of July 4th files
july4_files_upright = [
    'dataTorque/Upright-torque-Rod3-N2-angles.csv',
    'dataTorque/Upright-torque-Rod3-N3-moremagnets.csv',
    'dataTorque/Upright-torque-Rod3-N4-moremagnets.csv',
    'dataTorque/Upright-torque-Rod3-N5-moremagnets-furtherfrombridge.csv',
    'dataTorque/Upright-torque-Rod3-notangled.csv'
]

# List of July 11th files
july11_files_upright = [
    'New_dataTorque/Upright-torque-height-test_Rod3_N1.csv',
    'New_dataTorque/Upright-torque-height-test_Rod3_N2.csv',
    'New_dataTorque/Upright-torque-height-test_Rod3_N22.csv',
    'New_dataTorque/Upright-torque-height-test_Rod3_N3.csv',
    'New_dataTorque/Upright-torque-height-test_Rod3_N32.csv',
    'New_dataTorque/Upright-torque-height-test_Rod3_N4.csv',
    'New_dataTorque/Upright-torque-height-test_Rod3_N42.csv',
    'New_dataTorque/Upright-torque-height-test_Rod3_N12.csv'
]

compiled_data = []

# Parse metadata and extract values
for file_path in july11_files_upright:
    fname = os.path.basename(file_path)
    rod_match = re.search(r'Rod(\d+)', fname)
    rod_id = f"Rod{rod_match.group(1)}" if rod_match else "Unknown"
    trial_match = re.search(r'N\d+', fname)
    trial = trial_match.group(0) if trial_match else "N1"
    
    df = pd.read_csv(file_path, header=None)
    
    for i, row in df.iterrows():
        for idx, val in enumerate(row):
            compiled_data.append({
                "rod_id": rod_id,
                "trial": trial,
                "measurement_index": i,
                "point_index": idx,
                "value": val
            })

# Convert to DataFrame and save
compiled_df = pd.DataFrame(compiled_data)
compiled_path = "compiled_upright_torque_july11.dat"
compiled_df.to_csv(compiled_path, sep=",", index=False)
compiled_path
```
