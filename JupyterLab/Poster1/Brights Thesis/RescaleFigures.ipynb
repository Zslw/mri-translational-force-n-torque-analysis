```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plot
lengths = [6.3, 12.7, 19.0]  # Implant lengths in cm
heating_bipolar = [2/5, 1.8, 12.55]  # Max temp rise for bipolar leads
heating_unipolar = [2/4, 1.7, 6.8]  # Max temp rise for unipolar leads
heating_ti = [2/4, 1.5, 2/3]  # Max temp rise for titanium rods

# Error values for the plot
error_bipolar = [0.18, 0.3, 0.65]
error_unipolar = [0.04, 0.2, 0.35]
error_ti = [0.1, 0.2, 0.1]

# Create bar chart
x = np.arange(len(lengths))
width = 0.25

fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(x - width, heating_ti, width, yerr=error_ti, capsize=5, label='Ti', color='blue', alpha=0.8)
ax.bar(x, heating_bipolar, width, yerr=error_bipolar, capsize=5, label='Bipolar', color='orange', alpha=0.8)
ax.bar(x + width, heating_unipolar, width, yerr=error_unipolar, capsize=5, label='Unipolar', color='gray', alpha=0.8)

# Add labels, title, and legend
ax.set_xlabel("Implant Length (cm)", fontsize=14)
ax.set_ylabel("Maximum Temp. Rise (Â°C)", fontsize=14)
ax.set_title("Variation of Implant Heating with Length", fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels([f"{l} cm" for l in lengths], fontsize=12)
ax.legend(fontsize=12)
ax.grid(axis="y", linestyle="--", alpha=0.7)

# Save as high-resolution file
plt.tight_layout()
plt.savefig("/Users/danieljuarezluna/Documents/LaTex/LaTex--projects/SeminarFall2024 Poster/Figure_4_8_Implant_Heating_Length_With_Errors.pdf", format="pdf", dpi=300)
plt.show()
```
